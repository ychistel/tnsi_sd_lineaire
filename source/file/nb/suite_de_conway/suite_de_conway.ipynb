```python
def creer_file():
    return []

def est_vide(file):
    return file == []

def enfiler(file,elt):
    file.append(elt)
    return file

def defiler(file):
    if not est_vide(file):
        tete = file.pop(0)
    return tete
```

```python
def terme_en_file(terme):
    f1 = creer_file()
    for chiffre in terme:
        f1 = enfiler(f1, eval(chiffre))
    return f1
```

```python
def enumere_terme(f1):
    f2 = creer_file()
    c = defiler(f1)
    n = 1
    while not est_vide(f1):
        tete = defiler(f1)
        if c == tete:
            n += 1
        else:
            f2 = enfiler(f2,(n,c))
            c = tete
            n = 1
    f2 = enfiler(f2,(n,c))
    return f2
```

```python
def terme_suivant(f2):
    terme = ""
    while not est_vide(f2):
        t = defiler(f2)
        terme += str(t[0]) + str(t[1])
    return terme
```

```python
def suite_conway(n):
    suite = ['1']
    t = suite[0]
    for i in range(n-1):
        f1 = terme_en_file(t)
        f2 = enumere_terme(f1)
        t = terme_suivant(f2)
        suite.append(t)
    return suite
        
```

```python
suite_conway(20)
```
